```python
# 1. 税抜価格の入力を受け取る
# input()で受け取った文字列を、計算のためにint()で整数に変換します
price_withoutTax = input("税抜価格を入力してください：")
price_withoutTaxInt = int( price_withoutTax )

# 2. 税込価格を計算する (10%加算)
# 1.1をかけると10%増しの価格になります
price_withTax = price_withoutTaxInt * 1.1

# 3. 小数点以下を四捨五入する
# 0.5を足してからint()で整数（小数点以下を切り捨て）にすると、四捨五入と同じ結果になります
price_withTaxInt = int( price_withTax + 0.5)

# 4. 税込価格によって送料の処理を分ける (if-else文)
if price_withTaxInt >= 2000:
    # 2000円以上の場合
    print("送料は無料です")
    final_price = price_withTaxInt
else:
    # 2000円未満の場合
    print("送料として350円かかります")
    final_price = price_withTaxInt + 350

# 5. 最終的な送料込み価格を表示する
print(f"送料込みの価格は{final_price}円です。")
```
