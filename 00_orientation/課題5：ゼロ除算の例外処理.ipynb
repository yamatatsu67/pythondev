```python
# 問題文をtry： に入れる
try:
    a = int(input("割られる数を入力してください："))
    b = int(input("割る数を入力してください："))
    
    # 割り算を実行
    c = a / b

# ゼロで割り算した場合に出る例外を except: に入れる
except ZeroDivisionError:
    print("エラー：0で割り算しないでください")

# 数値ではない文字列が入力された場合に出る例外を ValueError: に入れる
except ValueError:
    print("エラー：数値（整数）を入力してください")

# 例外の発生が無い場合は else: で処理する
else:
    print(f"{a} ÷ {b} = {c}")

# 例外発生の有無に関わらずメッセージを表示する
finally:
    print("処理を終了します")
```
