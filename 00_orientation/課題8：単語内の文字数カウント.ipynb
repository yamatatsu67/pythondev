```python
# 空のリストを作成
WordList = []

# キーがアルファベット小文字の各文字で、値が0の辞書を作成
WordCountList0 = dict.fromkeys(list("abcdefghijklmnopqrstuvwxyz"), 0)

# 未入力でリターン（エンター）キーを押されるまで、英単語の入力の受け付け
while True:
    word = input("英単語を入力してください：")
    if word == "":
        break
    
# 入力された英単語を小文字に変換したものをリストに格納
    WordList.append(word.lower())

# 入力された英単語を構成する文字を1文字ずつ確認して小文字に変換し、登場した数を加算
    for alphabet in word.lower():
        if alphabet in WordCountList0:
           WordCountList0[alphabet] += 1

# 入力された英単語のリストをABC順に並べ替えてから一覧表示
WordList.sort()
print(f"入力した英単語：{WordList}")

#辞書を使って加算した値のうち、1文字以上あったすべての文字について「○が△個ありました」を表示
for key, values in WordCountList0.items():
    if values > 0:
        print(f"{key}が{values}個ありました")
```
