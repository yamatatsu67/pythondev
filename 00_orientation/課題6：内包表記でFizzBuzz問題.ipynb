```python
# 1から100までを num に代入しながらループ
# リスト内包にするとき num の range(1,101) は for で最後に持ってくる
# リストに内包するとき "判定表記" if 条件 else "判定表記" if 条件 を繰り返す
# 問題文は3で割り切れる数のときは"Fizz"から記載されているが、3と5で割り切れる数のときは"FizzBuzz"で判定したい
# 先に3と5で割り切れる数のときの"FizzBuzz"で判定を行いリストに格納させて、次から3で割り切れる数のときは"Fizz"の判定等を入れていく

fizzbuzz_list = ["FizzBuzz" if num % 3 == 0 and num % 5 == 0 else "Fizz" if num % 3 == 0 else "Buzz" if num % 5 == 0 else str(num) for num in range(1, 101)]

# 生成されたリストを全件出力
print(fizzbuzz_list)
```
