```python
#@title 九九のリストを作成する
#@title 九九の表を表示する
for v in range(1, 10):
    for h in range(1, 10):
        print(f"{(v * h):3d}", end="")

    print("")
```

```python
#@title 九九のリストを作成する
# リスト内包表記を使って 9x9 の2次元リストを作成する
# x (縦) が 1から9まで、 y (横) が 1から9まで
# この変数を KukuTable に格納する
# y (横) から range
KukuTable = [[x * y for y in range(1, 10)] for x in range(1, 10)]

#@title 九九の表を表示する (インデックスを使う方法)
# KukuTableから取り出して表示する、取り出すときに場所/住所の指示に注意
# printは教科書と同様に :3d, end="" を使用する
for X in range(0,9):
    for Y in range(0,9):
        xy = KukuTable[X][Y]

        print(f"{xy:3d}", end="")
    
    print("")
```
