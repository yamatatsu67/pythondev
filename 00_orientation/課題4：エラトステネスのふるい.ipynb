```python
# 1から100までを順番にチェックするループ
for num in range(1, 101):
    
    # 問題文記載の素数でないものを除外していく

    # 1は素数ではない
    if num == 1:
        continue
        
    # 2を除いた2の倍数は素数ではない
    if num > 2 and num % 2 == 0:
        continue  
        
    # 3を除いた3の倍数は素数ではない
    if num > 3 and num % 3 == 0:
        continue  
        
    # 5を除いた5の倍数は素数ではない
    if num > 5 and num % 5 == 0:
        continue  
        
    # 7を除いた7の倍数は素数ではない
    if num > 7 and num % 7 == 0:
        continue  
        
    print(num)
```
