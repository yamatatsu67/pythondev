```python
total_score = 0
count = 0

while True:
    score_INP = input("点数を入力してください：")
    score = int(score_INP)
    
    if score == -1:
        break
    
    # 入力された点数を表示する
    print(f"入力された点数： {score}")
    
    total_score += score
    count += 1

if count > 0:
    #平均点を計算する
    average = total_score / count
    
    #小数第2位を四捨五入し、小数第1位までの数値にする
    round_average = round(average, 1)
    
    #四捨五入した結果の変数を表示する
    print(f"{count}人のテストの平均点は{round_average}点です")
else:
    # 最初に-1が入力された場合
    # 区切り線とメッセージを表示する
    print("---") # 区切り線
    print("点数が入力されませんでした。")
```
