```python
# 命令1: 名前の入力を行いnameに格納
name = input("名前を入力してください：")

# 命令2: 現在の年齢の入力を行いageに格納
age = input("現在の年齢を入力してください：")

# 命令3: 来年の年齢を計算して表示
# input()で受け取った年齢は「文字列」なので、計算できるように「整数(int)」に変換後current_ageに格納
# 来年の年齢を計算しnext_ageに格納
current_age = int(age)
next_age = current_age + 1

# L2Ch6_4のf-string (f"...") を使い、{}の中に変数や計算結果を簡単に入れて表示します
print(f"私の名前は {name} です。来年で {next_age} 歳になります。")
```
